### 1.2 Nutrient rule filter system

```python
# nutrient rule system

rules = {
    "Protein": (0, 60),
    "Fat": (0, 100),
    "Carbohydrates": (0, 100),
}
```

#### I implemented a rule-based validation function to identify foods whose macronutrient and caloric values were not physiologically consistent with standard 100g nutritional reporting.

```python
def check_validity(row):
    issues = []
    for col, (low, high) in rules.items():
        if not (low <= row[col] <= high):
            issues.append(f"{col} out-of-range")
    return ", ".join(issues) if issues else "OK"
```

```python
df["validation"] = df.apply(check_validity, axis=1)
# # pd.set_option("display.max_rows", None)
df[df["validation"] != "OK"]
```

```python
invalid_foods = df.loc[df["validation"] != "OK", "food"]
print(invalid_foods.tolist())
```

#### I would say there are these type of anomalies: Meats or fish extremely high in protein, flours extremely high in carbs, Sweats and sauces

#### What do they have in common? They probably are a different type of serving

```python
# Save the invalid foods to a CSV file for further review
different_serving = df[df["validation"] != "OK"]
different_serving.to_csv("different_serving_foods.csv", index=False)
```
